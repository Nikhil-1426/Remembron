```python
#importing necessary modules
import speech_recognition as sr
import os
import calendar
r = sr.Recognizer()

#For the list of all available devices you can use: sr.Microphone.list_microphone_names()

#Microphones according to the need of the user
blu_mic = sr.Microphone(device_index=2)
mic = sr.Microphone()

#Keyword Arrays to form the reminders
default_keywords = ["meet","remind","remember","reminder","medicine","medication","meeting","appointment","meetup","get together","pills","therapy","medical","pill"] #Absolute 
digits = ['1:00','2:00','3:00','4:00','5:00','6:00','7:00','8:00','9:00','10:00','11:00','12:00']
AP = ["a.m.","a.m","p.m.","p.m"]
time_keywords = ["tomorrow","today","afternoon","noon","evening","tonight","lunch","morning"]
personal_keywords = [] #will be given by the user to save subjects which he wants to be transcribed especially in a seperate folder. i.e, topics important or needed for them.

usr_connection = "bluetooth" #Whether the user is using a wireless headset or the default microphone/wired headset for audio input

# Create a function to write recognized speech to a text file
def write_to_file(text):
    filename = "recognized_speech.txt"
    if not os.path.exists(filename):
        open(filename, "w").close()
    with open(filename, "a") as f:
        f.write(text + "\n")

# Create a function to write reminder transcript(The speech content containing information pertaining the reminders which we need to make) to a text file
def writen_to_file(text):
    filename = "reminder_transcript.txt"
    if not os.path.exists(filename):
        open(filename, "w").close()
    with open(filename, "a") as f:
        f.write(text + "\n")    

# Create a function to write speech pertining personal keywords to a text file
def perlogs(text):
    filename = "personal_logs.txt"
    if not os.path.exists(filename):
        open(filename, "w").close()
    with open(filename, "a") as f:
        f.write(text + "\n")    
               
#Audio Recording depending on if user is connected to bluetooth headset or not.
while True:
    if usr_connection == "bluetooth":
        z=0
        for i in range(0,1):   # Commencing the audio recording loop 
            with blu_mic as source:
                print("Recording... ")
                r.adjust_for_ambient_noise(source)
                audio = r.record(source, duration = 10)
                st = r.recognize_google(audio)
                print(st)   #The audio detected and displayed
            z=z+1
            if z==100:
                break
        break
    elif usr_connection == "default":
        z=0
        for i in range(0,1):   #Commencing the audio recording loop 
            with mic as source:
                print("Recording... ")
                r.adjust_for_ambient_noise(source)
                audio = r.record(source, duration = 10)
                st = r.recognize_google(audio)
                print(st)   #The audio detected and displayed
            z=z+1
            if z==50:
                break
        break
    else:
        break
#Saving the audio transcript in the general file
write_to_file(st)

#Array conversion of the audio.
split = st.split(" ")

#Addition of the prominent keywords if present in the file to the reminder transcript
default_keywords_list = []
digits_list = []
time_keywords_list = []
AP_list = []


#Creating the seperate parameters needed for saving the reminder.
for x in split:
    if x in default_keywords:
        if x not in default_keywords_list:
            default_keywords_list.append(x)            # Reason
    elif x in digits:
        if(len(digits_list)<1):      
            digits_list.append(x)                      # Time
    elif x in AP:
        if(len(AP_list)<1):
            AP_list.append(x)                          # a.m. or p.m.
    elif x in time_keywords:
        if x not in time_keywords_list:
            time_keywords_list.append(x)               #Date or when
print(default_keywords_list)
print(digits_list)
print(time_keywords_list)
txt = digits_list+AP_list+default_keywords_list+time_keywords_list
txt = " ".join(txt)
writen_to_file(txt) #adding reminder point to the reminder_transcript file.
txt = [] #resetting txt 

#Updating Personal logs
pr_txt = []

for x in split:
    if x in personal_keywords:
        pr_txt = " ".join(pr_txt)

perlogs(pr_txt) #Appending the personal_logs file.

```
