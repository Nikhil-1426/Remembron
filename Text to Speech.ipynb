```python
import schedule
import time
import datetime


with open("reminder_transcript.txt",'r+') as file:
    lines = file.readlines()
    for line in lines:
        #Reminder TTS working:
        reminder = line
        reminder=reminder.split()
        reminder[1]=reminder[1][0]+reminder[1][2]
        reminder=' '.join(reminder)
        print(reminder)


        # Function to remind the user
        def remind(reminder):
            print(f"It's time for your {reminder.split()[0]}")

        # Function to parse reminder strings and set up reminders
        def setup_reminders():

                parts = reminder.split()
                time_str = parts[0] + ' ' + parts[1]
                activity = ' '.join(parts[2:])

                # Parse time
                reminder_time = datetime.datetime.strptime(time_str, '%I:%M %p')

                # Check if reminder is for today or tomorrow
                if parts[-1] == 'today':
                    schedule.every().day.at(reminder_time.strftime('%H:%M')).do(remind, activity)
                elif parts[-1] == 'tomorrow':
                    tomorrow = datetime.date.today() + datetime.timedelta(days=1)
                    reminder_date = datetime.datetime.combine(tomorrow, reminder_time.time())
                    schedule.every().day.at(reminder_date.strftime('%H:%M')).do(remind, activity)

        # Start scheduling reminders
        setup_reminders()

        # Continuously check and execute scheduled reminders
        while True:
            schedule.run_pending()
            time.sleep(45)
```
